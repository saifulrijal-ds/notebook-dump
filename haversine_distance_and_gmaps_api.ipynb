```python
# concatenating lat and long to create a consolidatedlocationas accepted by haversine function
cust_loc_df["coor"] = tuple(zip(cust_loc_df["Lat"], cust_loc_df["Lon"]))
```

```python
hotel_loc_df["coor"] = list(zip(hotel_loc_df["Lat"], hotel_loc_df["Lon"]))
```

```python
hotel_loc_df.head()
```

```python
hotel_loc_df.info()
```

```python
# defining a function to calculate distance between two locations
# loc1 = location of exixtsing hotel
# loc2 = location of customer

def distance_from(loc1, loc2):
    dist=hs.haversine(loc1, loc2)
    return round(dist, 2) # result in km
```

```python
# running a loop which will parse customers location one by one to distance from function
for _, row in hotel_loc_df.iterrows():
    cust_loc_df[row["Hotel"]] = cust_loc_df["coor"].apply(lambda x: distance_from(row["coor"], x))
```

```python
cust_loc_df
```

```python
# Mapping through folium

def create_map(cust, hotel):
    m = folium.Map(location=[cust["Lat"].mean(), cust["Lon"].mean()], zoom_start=12, tiles="OpenStreetMap")
    
    for _, row in cust.iterrows():
        
        if row["Condition"]=="Out":
            cluster_color="blue"
        else:
            cluster_color="green"
            
        folium.CircleMarker(
            location=[row["Lat"], row["Lon"]],
            radius=5, # in pixel
            popup=row[["Condition", "Location"]],
            color=cluster_color,
            fill=True,
            fill_color="red"
        ).add_to(m)
        
    for _, row in hotel.iterrows():
        folium.Marker(
            location=[row["Lat"], row["Lon"]],
            radius=5,
            popup=row["Hotel"],
            color="red",
            fill=True,
            fill_color="red"
        ).add_to(m)
        
    for _, row in hotel.iterrows():
        folium.Circle(
            location=[row["Lat"], row["Lon"]],
            radius=3000, # in meters
            popup=row["Hotel"],
            color="black",
            fill=False,
        ).add_to(m)
        
    return m
```

Please attention about radius in CircleMarker https://stackoverflow.com/questions/50954840/displaying-radius-in-meters-with-folium

```python
hotel_list = ["A", "B", "C", "D", "E", "F", "G", "H"]
```

```python
cust_loc_df["Condition"] = "Out"

# change to "In" if less than 3 Km
for i in hotel_list:
    cust_loc_df["Condition"][cust_loc_df["Condition"]=="Out"] = cust_loc_df[cust_loc_df["Condition"]=="Out"][i].apply(lambda x: "In" if x < 3 else "Out")
```

```python
cust_loc_df
```

```python
m = create_map(cust_loc_df, hotel_loc_df)
```
